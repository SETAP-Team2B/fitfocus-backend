```python
food_nutrient["nutrient"] = food_nutrient["nutrient"].replace(
    nutrient["name"].values,
    [
    "calories",
    "fat_g",
    "saturates_g",
    "trans_fat_g",
    "protein_g",
    "salt_g",
    "fiber_g",
    "carbohydrates_g",
    "sugars_g",
    "cholesterol_g",
    ]
)
```

```python
food_list = []

for food_name in food_nutrient["food"].unique():
    macros_list = food_nutrient[food_nutrient["food"] == food_name].drop("food", axis=1).to_dict("records")
    final_macros = {}

    # scales because 
    portion_scaler = 1 if food_name not in food_portion["food"].values else food_portion[food_portion["food"] == food_name]["gram_weight"].values[0] / 100.0

    for macro in macros_list:
        if macro["nutrient"] == "calories":
            calories = macro["amount"] * portion_scaler
        else:
            final_macros[macro["nutrient"]] = round(macro["amount"] * portion_scaler, 2) # round the macro to 2 decimal places, no need for any more precision, 2 is already extremely precise

    # default sample serving size is 100g, gets scaled if a portion is found in the dataset
    food_dict = {
        "name": food_name,
        "sample_size": 100 if food_name not in food_portion["food"].values else food_portion[food_portion["food"] == food_name]["amount"].values[0],
        "sample_units": "g" if food_name not in food_portion["food"].values else food_portion[food_portion["food"] == food_name]["modifier"].values[0],
        "sample_calories": int(calories),
        "sample_macros": final_macros,
    }
    
    food_list.append(food_dict)

print(len(food_list))
```
